```python
import pandas as pd
import numpy as np
from datetime import date
```

```python
dfs = pd.read_csv('./cycle-share-dataset/station.csv')
dfs.head()
```

```python
dft = pd.read_csv('./cycle-share-dataset/trip.csv', error_bad_lines=False)
dft = dft.drop_duplicates()
dft.head()
```

```python
dfw = pd.read_csv('./cycle-share-dataset/weather.csv')
dfw.head()
```

## I. What is the average trip duration for a borrowed bicycle?

```python
trip_dur = np.average(dft['tripduration'])
trip_dur
```

## II. What's the most common age of a bicycle-sharer?

```python
trip_age = np.mean(dft['birthyear'])
date.today().year - trip_age
```

### III. Given all the weather data here, find the average precipitation per month, and the median precipitation.

Average

```python
dfw['DATE'] = pd.to_datetime(dfw['Date'], errors='coerce')
dfw['MONTH'] = dfw['DATE'].dt.month
wea_avg = pd.DataFrame(dfw['Precipitation_In'].groupby(dfw.MONTH).mean())
wea_avg
```

Median

```python
wea_med = pd.DataFrame(dfw['Precipitation_In']).groupby(dfw.MONTH).median()
wea_med
```

## IV. What’s the average number of bikes at a given bike station?

```python
stat_bike = pd.DataFrame(dfs['current_dockcount']).groupby(dfs.station_id).mean()
stat_bike
```

## V. When a bike station is modified, is it more likely that it’ll lose bikes or gain bikes? How do you know?

```python
inst_count = np.average(dfs['install_dockcount'])
mod_count = np.average(dfs['current_dockcount'])
if(inst_count > mod_count):
    print('More likely to lose bikes.')
elif(inst_count < mod_count):
    print('More likely to gain bikes.')
else:
    print('Neither gain nor lose.')
```

## Address with most bikes

```python
add_bikes = pd.DataFrame(dfs[['name', 'current_dockcount']]).sort_values('current_dockcount', ascending=False).head(5)
add_bikes
```

## Most popular station to borrow from

```python
pop_stat = pd.DataFrame(dft['from_station_name'].groupby(dft.from_station_name).count()).sort_values('from_station_name', ascending=False).head(5)
pop_stat
```
